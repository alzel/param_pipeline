```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from plotnine import *
import glob
import os
```

```python
## create data frame with combinations of hyperparameters and results
# 2018_8_3: update to read all lines=epochs available
# 2018_8_15: make function and time

def read_pipeline_results_V1(name = 'results/data1/*.csv',
                         cols_par=['alpha','beta1','beta2','epsilon','decay']):

    cols_res = ['epoch','loss_train','r2_train','loss_val','r2_val','loss_test','r2_test']
    filelist = glob.glob(name)
    results = pd.DataFrame(np.nan, index=range(1), dtype='float32', 
                      columns=cols_par+cols_res)
    results_df = pd.DataFrame(np.nan, index=range(1), dtype='float32', 
                      columns=cols_par+cols_res)

    k=0
    for i in filelist:
        with open(i, 'r') as f:
            fin = f.readlines()[1:]
            f.close()
            for ii in range(len(fin)):
                results.iloc[0,0] = float(i.split('/', 1)[1].split('_')[3])
                results.iloc[0,1] = float(i.split('/', 1)[1].split('_')[4])
                results.iloc[0,2] = float(i.split('/', 1)[1].split('_')[5])
                results.iloc[0,3] = float(i.split('/', 1)[1].split('_')[6])
                results.iloc[0,4] = float(i.split('/', 1)[1].split('_')[7])
                # results.iloc[0,5] = float(i.split('/', 1)[1].split('_')[8]) # dropout fixed
                # results.iloc[0,6] = float(i.split('/', 1)[1].split('_')[9].split('.')[0]) # batch fixed
                # results.iloc[0,7] = str(i.split('/', 1)[1].split('_')[2]) # model fixed
                results.iloc[0,5] = ii+1
                results.iloc[0,6] = float(fin[ii].split(',')[3])
                results.iloc[0,7] = float(fin[ii].split(',')[1])
                results.iloc[0,8] = float(fin[ii].split(',')[6])
                results.iloc[0,9] = float(fin[ii].split(',')[4])
                if ii==len(fin)-1:
                    if os.path.exists(i+'_testset'):
                        with open(i+'_testset', 'r') as f2:
                            fin2 = f2.readlines()[1:]
                            f2.close()
                            results.iloc[0,10] = float(fin2[0].split(',')[0])
                            results.iloc[0,11] = float(fin2[0].split(',')[1])
                results_df = results_df.append(results, ignore_index=True)
                results = pd.DataFrame(np.nan, index=range(1), dtype='float32', 
                                      columns=cols_par+cols_res)

    results_df = results_df.drop(0)
    results_df.reset_index(drop=True, inplace=True)
    
    return results_df
```

```python
import time
t = time.time()
# do stuff
results = read_pipeline_results_V1()
elapsed = time.time() - t
print(elapsed)
```

```python
## create data frame with combinations of hyperparameters and results
# 2018_8_3: update to read all lines=epochs available
# 2018_8_15: update to read_csv function for speed, make function

def read_pipeline_results_V2(name = 'results/data1/*.csv',
                         cols_par=['alpha','beta1','beta2','epsilon','decay']):

    # define stuff
    cols_res = ['epoch','loss_train','r2_train','loss_val','r2_val','loss_test','r2_test']
    results = pd.DataFrame(np.nan,index=range(1),dtype='float32',columns=cols_par+cols_res)
    filelist = glob.glob(name)

    for i in filelist:
       # reset frames 
        results_tmp1 = pd.DataFrame(np.nan,index=range(1),dtype='float32',columns=cols_par)

        # read parameters
        results_tmp1.iloc[0,0] = float(i.split('/', 1)[1].split('_')[3])
        results_tmp1.iloc[0,1] = float(i.split('/', 1)[1].split('_')[4])
        results_tmp1.iloc[0,2] = float(i.split('/', 1)[1].split('_')[5])
        results_tmp1.iloc[0,3] = float(i.split('/', 1)[1].split('_')[6])
        results_tmp1.iloc[0,4] = float(i.split('/', 1)[1].split('_')[7])
        # results_tmp1.iloc[0,5] = float(i.split('/', 1)[1].split('_')[8]) # dropout fixed
        # results_tmp1.iloc[0,6] = float(i.split('/', 1)[1].split('_')[9].split('.')[0]) # batch fixed
        # results_tmp1.iloc[0,7] = str(i.split('/', 1)[1].split('_')[2]) # model fixed

        # read results
        results_tmp2 = pd.read_csv(i)
        results_tmp2['epoch'] = results_tmp2['epoch']+1
        results_tmp2.rename(columns={"loss":"loss_train","coef_det_k":"r2_train","val_loss":"loss_val","val_coef_det_k":"r2_val"},inplace=True)
        results_tmp2.drop(['corr_coef_k','val_corr_coef_k'],axis=1,inplace=True)

        # reshape, concatenate, append
        results_tmp1 = pd.concat([results_tmp1]*results_tmp2.shape[0],ignore_index=True)
        results = results.append(pd.concat([results_tmp1,results_tmp2],axis=1))

        # read .csv_testset, always add to last one
        results_tmp3 = pd.read_csv(i+'_testset')
        results_tmp3.rename(columns={"MSE":"loss_test","coef_determination":"r2_test"},inplace=True)
        results_tmp3.drop(['corr_coef','corr_coef_plot'],axis=1,inplace=True)
        results.iloc[-1,6] = results_tmp3['loss_test'].values
        results.iloc[-1,9] = results_tmp3['r2_test'].values

        # clean
        results = results.reset_index(drop=True)

    # clean out of loop
    return results.drop(0).reset_index(drop=True)
```

```python
import time
t = time.time()
# do stuff
results2 = read_pipeline_results_V2()
elapsed = time.time() - t
print(elapsed)
```
